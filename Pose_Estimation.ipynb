```python
import cv2 
```

```python
import mediapipe as mp
```

```python
import pymysql
import pandas as pd
import time
import datetime
```

```python
conn=pymysql.connect(host='localhost',port=int(3306),user='root',passwd='',db='cv')
#conn=pymysql.connect(host='192.168.106.50',port=int(3306),user='iot2',passwd='',db='dashboard_main')
```

```python
cap = cv2.VideoCapture(0)
```

```python
mp_pose = mp.solutions.pose
mp_draw = mp.solutions.drawing_utils
ts = time.time()
timestamp = datetime.datetime.fromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S')
pose = mp_pose.Pose()
cursor = conn.cursor()
```

```python
while True:
    ret,frame = cap.read()
    flipped = cv2.flip(frame,flipCode=1)
    frame1 = cv2.resize(flipped,(640,480))
    rgb_img = cv2.cvtColor(frame1, cv2.COLOR_BGR2RGB)
    result = pose.process(rgb_img)
    image_width = 640
    image_height = 480
    xpos = result.pose_landmarks.landmark[mp_pose.PoseLandmark.NOSE].x * image_width
    ypos = result.pose_landmarks.landmark[mp_pose.PoseLandmark.NOSE].y * image_height
    
    try:
        print('X position of nose is', result.pose_landmarks.landmark[mp_pose.PoseLandmark.NOSE].x * image_width)
        print('Y position of nose is', result.pose_landmarks.landmark[mp_pose.PoseLandmark.NOSE].y * image_height)
        
        if ypos > 300:
            stat = 0
    
        else:
            stat = 1
            
        sql = "INSERT INTO `detection` (`id`, `x`, `y`, `z`, `status`, `timestamp`) VALUES (%s, %s, %s, %s, %s, %s)"
        cursor.execute(sql, ('NULL',xpos,ypos,1,stat,timestamp))
        conn.commit()
                       
    except:
        pass
    
    mp_draw.draw_landmarks(frame1, result.pose_landmarks, mp_pose.POSE_CONNECTIONS)
    cv2.imshow("frame", frame1)
    
    key = cv2.waitKey(1) & 0xFF
    if key ==ord("q"):
        conn.close()
        break
```
